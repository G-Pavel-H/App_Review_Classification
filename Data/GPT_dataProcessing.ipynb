```python
df3 = pd.read_csv("updated_reviews_labeled.csv")
df3.head()
```

```python
def add_new_labels(label):
    if label == "bug report":
        return 0
    elif label == "feature request":
        return 1
    elif label == "rating":
        return 2
    else:
        return 3
```

```python
df3['new_label'] = df3['label'].apply(add_new_labels)
df3.head()
```

```python
num_BR = df3.loc[df3["new_label"] == 0].shape[0]
num_RT = df3.loc[df3["new_label"] == 2].shape[0]
num_FR = df3.loc[df3["new_label"] == 1].shape[0]
num_UE = df3.loc[df3["new_label"] == 3].shape[0]

print(num_BR)
print(num_RT)
print(num_FR)
print(num_UE)
```

```python
desired_rows_per_label = {
    0: 5000,
    1: 5000,
    2: 5000,
    3: 5000
}

balanced_df = pd.DataFrame(columns=df3.columns)

# Loop through each label and balance the rows
for label in [0, 1, 2, 3]:
    label_df = df3[df3['new_label'] == label]
    label_df = label_df.sort_values(by='review', ascending=False, key=lambda x: x.str.len())
    label_df = label_df[:desired_rows_per_label[label]]
    balanced_df = pd.concat([balanced_df, label_df])

gpt_balanced_20000 = balanced_df
```

```python
num_BR = gpt_balanced_20000.loc[gpt_balanced_20000["new_label"] == 0].shape[0]
num_RT = gpt_balanced_20000.loc[gpt_balanced_20000["new_label"] == 2].shape[0]
num_FR = gpt_balanced_20000.loc[gpt_balanced_20000["new_label"] == 1].shape[0]
num_UE = gpt_balanced_20000.loc[gpt_balanced_20000["new_label"] == 3].shape[0]

print(num_BR)
print(num_RT)
print(num_FR)
print(num_UE)
```

```python
import pandas as pd
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
nltk.download('punkt')  # For tokenization


# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Function to lemmatize a sentence
def lemmatize_sentence(sentence):
    # Tokenize the sentence
    tokens = word_tokenize(sentence)
    # Lemmatize each word in the sentence
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    # Join the lemmatized tokens back into a sentence
    return ' '.join(lemmatized_tokens)

# Apply the lemmatization function to the desired column
gpt_balanced_20000['lemmatization'] = gpt_balanced_20000['review'].apply(lemmatize_sentence)

# Show the DataFrame
gpt_balanced_20000.head()
```

```python
import pandas as pd
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import nltk


# Example DataFrame

# Initialize the PorterStemmer
stemmer = PorterStemmer()

# Function to stem a sentence
def stem_sentence(sentence):
    # Tokenize the sentence
    tokens = word_tokenize(sentence)
    # Stem each word in the sentence
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    # Join the stemmed tokens back into a sentence
    return ' '.join(stemmed_tokens)

# Apply the stemming function to the desired column
gpt_balanced_20000['stemming'] = gpt_balanced_20000['review'].apply(stem_sentence)

# Show the DataFrame
gpt_balanced_20000.head()
```

```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# Load the list of stopwords in English
stop_words = set(stopwords.words('english'))



# Function to remove stopwords from a sentence
def remove_stopwords(sentence):
    # Tokenize the sentence
    tokens = word_tokenize(sentence)
    # Remove stopwords from the tokens list
    filtered_tokens = [word for word in tokens if not word.lower() in stop_words]
    # Join the filtered tokens back into a sentence
    return ' '.join(filtered_tokens)

# Apply the stopword removal function to the desired column
gpt_balanced_20000['no_stopwords'] = gpt_balanced_20000['review'].apply(remove_stopwords)

# Show the DataFrame
gpt_balanced_20000.head()
```
