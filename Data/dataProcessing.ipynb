```python
import nltk
import pandas as pd
```

```python
df_intial = pd.read_csv('all_combined_labeled.csv')
df_intial.head()
```

```python
df_kaggle = pd.read_csv('dataset_balanced_8000.csv')  
df_kaggle.head()
```

```python
df_kaggle_md = df_kaggle[['review', 'label', 'new_label']].copy()
df_kaggle_md.head()
```

```python
df_intial_md = df_intial.replace('PD', 'bug report')
df_intial_md = df_intial_md.replace('RT', 'rating')
df_intial_md = df_intial_md.replace('UE', 'user experience')
df_intial_md = df_intial_md.replace('FR', 'feature request')
df_intial_md.head()

```

```python
def add_new_labels(label):
    if label == "bug report":
        return 0
    elif label == "feature request":
        return 1
    elif label == "rating":
        return 2
    else:
        return 3
```

```python
df_intial_md['new_label'] = df_intial_md['label'].apply(add_new_labels)
```

```python
df_intial_md.head()
```

```python
final_data = pd.concat([df_intial_md, df_kaggle_md], axis=0) 
```

```python
num_BR = final_data.loc[final_data["new_label"] == 0].shape[0]
num_RT = final_data.loc[final_data["new_label"] == 2].shape[0]
num_FR = final_data.loc[final_data["new_label"] == 1].shape[0]
num_UE = final_data.loc[final_data["new_label"] == 3].shape[0]

print(num_BR)
print(num_RT)
print(num_FR)
print(num_UE)

final_data_unbalanced = final_data
```

```python
final_data_balanced = final_data

desired_rows_per_label = {
    0: 4000,
    1: 4000,
    2: 4000,
    3: 4000
}

balanced_df = pd.DataFrame(columns=final_data_balanced.columns)

# Loop through each label and balance the rows
for label in [0, 1, 2, 3]:
    label_df = final_data_balanced[final_data_balanced['new_label'] == label]
    label_df = label_df.sort_values(by='review', key=lambda x: x.str.len())
    label_df = label_df[:desired_rows_per_label[label]]
    balanced_df = pd.concat([balanced_df, label_df])

final_data_balanced = balanced_df
```

```python
num_BR = final_data_balanced.loc[final_data_balanced["new_label"] == 0].shape[0]
num_RT = final_data_balanced.loc[final_data_balanced["new_label"] == 2].shape[0]
num_FR = final_data_balanced.loc[final_data_balanced["new_label"] == 1].shape[0]
num_UE = final_data_balanced.loc[final_data_balanced["new_label"] == 3].shape[0]

print(num_BR)
print(num_RT)
print(num_FR)
print(num_UE)
```

```python
final_data_balanced.to_excel('final_data_balanced_16000.xlsx', index=False)
```

```python
final_data_unbalanced.to_excel(f"final_data_unblanced_{final_data_unbalanced.shape[0]}.xlsx", index=False)
```
